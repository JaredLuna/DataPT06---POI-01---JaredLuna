```python

#Una vez importados nuestros datos comenzamos con la limpieza de los mismos como mencionamos en la EDA
#Para el primer DataFrame (df1) vamos a eliminar todas las filas que contengan valores NA
df1_clean = df1.dropna()
df1_clean.head()

#Ya no tenemos valores vacios o NA
```

```python
Output_steam_df = df1_clean
Output_steam_df = Output_steam_df.drop('Unnamed: 0', axis= 1)
Output_steam_df = Output_steam_df.drop('url', axis= 1)
Output_steam_df = Output_steam_df.drop('reviews_url', axis= 1)
Output_steam_df = Output_steam_df.drop('app_name', axis= 1)

Output_steam_df.head()
```

```python
#Una ultima limpieza para las columnas...
def clean(x):
    x = x.replace('[', '')
    x = x.replace(']', '')
    x = x.replace("'", "")
    x = x.split(',')
    return x

Output_steam_df['genres'] = Output_steam_df['genres'].apply(clean)
Output_steam_df['tags'] = Output_steam_df['tags'].apply(clean)
Output_steam_df['specs'] = Output_steam_df['specs'].apply(clean)
Output_steam_df.head()
```

```python
#Una vez eliminados los valores faltantes guardamos neustra DF en un nuevo csv
df1_clean.to_csv('./Csvs/Output_steam_post_etl.csv')
```

```python
df2.head()
```

```python
#Para el segundo DataFrame (df2) vamos a eliminar la columna user_url y a desanidar la información de la última columna
df2 = df2.drop('user_url', axis= 1)
df2.head()
#Ya eliminamos la columna sobrante, ahora vamos a desanidar la info de la ultima columna
```

```python
#Tambien eliminamos la columna de items ya que la podemos relacionar con nuestra otra de datos expandidos
df2 = df2.drop('items', axis= 1)
```

```python
#Ya tenemos nuestros datos listos, así que procedemos a guardarlos en un nuevo csv
df2.to_csv('./Csvs/Australian_items_post_etl.csv')
```
